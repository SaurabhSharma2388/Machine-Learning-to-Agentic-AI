# Retrieval Augmented Generation (RAG)
In this lesson, Retrieval Augmented Generation (RAG) is the process of optimizing the output of a large language model , so it reference an authoritative knowledge base outside of its training data source before generating a response.



```python
!pip install groq
```

```python
import os
from groq import Groq
```

```python
client = Groq(api_key='API_KEY')
```

```python
data = "Mid-century office chair, Made in Italy, Aluminum base, 5 wheels."
prompt = f"Write a one-paragraph marketing description for this: {data}"
```

```python
completion = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": prompt}]
)
```

```python
print(completion.choices[0].message.content)
```

# Converting for RAG (Retrieval-Augmented Generation)

```python
from groq import Groq
```

```python
knowledge_base = {
    "chair": "Mid-century office chair, Made in Italy, Aluminum base, 5 wheels.",
    "desk": "Solid oak standing desk, motorized height adjustment, cable tray."
}
```

```python
query = "chair"
context = knowledge_base.get(query, "No information found.")
```

```python
rag_prompt = f"""
Use the following Context to answer the User Question.
Context: {context}
User Question: Write a short marketing description.
"""
```

```python
completion = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[{"role": "user", "content": rag_prompt}]
)
```

```python
print(completion.choices[0].message.content)
```
