```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

documents = [
    "The capital of France is Paris.",
    "Pyramids are ancient masonry structures located in Egypt.",
    "The Great Wall of China is a series of fortifications.",
    "The Eiffel Tower is located in Paris, France.",
    "A vector is a quantity having direction as well as magnitude."
]

print("Encoding documents into vectors...")
document_embeddings = model.encode(documents)
```

```python
user_query = "Tell me about famous landmarks in France."
query_embedding = model.encode([user_query])
```

```python
similarities = cosine_similarity(query_embedding, document_embeddings)
```

```python
best_index = np.argmax(similarities)
confidence_score = similarities[0][best_index]
```

```python
print(f"\nQuery: {user_query}")
print(f"Retrieved Context: {documents[best_index]}")
print(f"Similarity Score: {confidence_score:.4f}")

print(f"\nExample Vector (First 5 numbers of {len(document_embeddings[0])}):")
print(document_embeddings[0][:5])
```

# Sentiment Analysis

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('all-MiniLM-L6-v2')

categories = ["I am very happy and satisfied with this experience!",
              "I am very disappointed and upset with this service."]
category_names = ["Positive", "Negative"]
```

```python
category_vectors = model.encode(categories)

new_reviews = [
    "The food was delicious and the staff were incredibly kind.",
    "It took forever to arrive and it was broken when I opened it.",
    "Not the best, but it gets the job done."
]
```

```python
print("--- Sentiment Analysis Results ---")

for review in new_reviews:
    review_vector = model.encode([review])

    similarities = cosine_similarity(review_vector, category_vectors)[0]

    best_idx = similarities.argmax()
    sentiment = category_names[best_idx]
    score = similarities[best_idx]

    print(f"Review: \"{review}\"")
    print(f"Classified as: {sentiment} ({score:.4f} match)\n")
```
